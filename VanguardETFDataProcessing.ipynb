**DATA FORMATTING:**

```python
df = pd.read_csv("etfs.csv")
```

```python
df.head(100)
```

```python
df["Expense Ratio"] = df["Expense Ratio"].str.rstrip("%").astype(float)

def remove_dashes(percent):
  try:
    return float(percent)
  except:
    return None

df["YTD % Return"] = df["YTD % Return"].str.rstrip("%").apply(remove_dashes)
df["1 Year % Return"] = df["1 Year % Return"].str.rstrip("%").apply(remove_dashes)
df["3 Year % Return"] = df["3 Year % Return"].str.rstrip("%").apply(remove_dashes)
df["5 Year % Return"] = df["5 Year % Return"].str.rstrip("%").apply(remove_dashes)
df["10 Year % Return"] = df["10 Year % Return"].str.rstrip("%").apply(remove_dashes)
df["% Return Since Inception"] = df["% Return Since Inception"].str.rstrip("%").apply(remove_dashes)
```

```python
def format_date(input_date_string):
  return datetime.datetime.strptime(input_date_string, '%m/%d/%Y').date()

df['Inception Date'] = df['Inception Date'].apply(format_date)
```

```python
df["25-day Avg Volume"] = df["25-day Avg Volume"].str.replace(',', '').astype(int)
```

```python
df["50-day Avg Volume"] = df["50-day Avg Volume"].str.replace(',', '').astype(int)
```

```python
def remove_x(percent):
  try:
    return float(percent.rstrip("x"))
  except:
    return None

df["P/E Ratio"] = df["P/E Ratio"].apply(remove_x)
```

```python
def convert_to_dollars(money_amount):
  money_amount = money_amount.lstrip("$")
  if money_amount.endswith("M"):
    return float(money_amount.rstrip(" M")) * 1000000
  elif money_amount.endswith("B"):
    return float(money_amount.rstrip(" B")) * 1000000000
  else:
    return float(money_amount.rstrip(" T")) * 1000000000000

df["Fund Total Net Assets"] = df["Fund Total Net Assets"].apply(convert_to_dollars)
```

```python
df.rename(columns = {'Expense Ratio':'Expense Ratio (as a %)', 'Market Price':'Market Price (in $)', 'Fund Total Net Assets':'Fund Total Net Assets (in $)'}, inplace = True)
```

```python
df.head(100)
```

**DATA PROCESSING:**

```python
#because I'm not looking to invest in etfs with higher than 0.1% fee
df = df.loc[df['Expense Ratio (as a %)'] <= 0.1]
```

```python
#because I was advised that high risk investments generally outperform lower risk ones over the long run, and I'm planning for long term investment
df = df.loc[df['Risk/Reward Value'] >= 4]
```

```python
#because I already invest heavily in large indices, so I'm looking to diversify here
df = df.loc[df['Category'].str.contains("Large") == False]
```

```python
#because I am looking for the etfs that I invest in to be at least 5 years old, so there is some established credbility
df = df.loc[df["Inception Date"] <= datetime.date(datetime.date.today().year - 5, datetime.date.today().month, datetime.date.today().day)]
```

```python
#because I would like the etfs that I invest in to have performed at or above average during the past 5 years
df = df.loc[df["5 Year % Return"] >= df["5 Year % Return"].median()]
```

```python
#because I would like to pay less for greater earnings100
df = pd.concat([df.loc[df["P/E Ratio"].isnull()], df.loc[df["P/E Ratio"] < df["P/E Ratio"].median()]])
```
